# 03 - Prong 2: DGL R-GCN with Temporal Decay

Captures behavioral drift that static PyG embeddings miss.


```python
import torch
from bgi_trident.graph.dgl.model import compute_recency_decay

days = torch.tensor([0., 7., 14., 21., 28.])
for hl in [7, 14, 28]:
    decay = compute_recency_decay(days, half_life_days=hl)
    print(f'Half-life {hl}d: {[round(d.item(), 3) for d in decay]}')
```

A user who stopped ordering 3 weeks ago: PyG sees 0.91, DGL sees 0.43. The ensemble resolves this.
