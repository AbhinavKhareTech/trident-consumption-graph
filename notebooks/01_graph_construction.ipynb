# 01 - Heterogeneous Graph Construction

Build the consumption graph from synthetic Bangalore data. 6 node types, 8 cross-domain edge types.


```python
import pandas as pd
from pathlib import Path
DATA = Path('../src/data')
users = pd.read_csv(DATA / 'bangalore_users.csv')
restaurants = pd.read_csv(DATA / 'bangalore_restaurants.csv')
products = pd.read_csv(DATA / 'bangalore_products.csv')
venues = pd.read_csv(DATA / 'bangalore_venues.csv')
print(f'Users: {len(users)}, Restaurants: {len(restaurants)}, Products: {len(products)}, Venues: {len(venues)}')
```

```python
food_orders = pd.read_csv(DATA / 'interactions/food_orders.csv')
instamart_orders = pd.read_csv(DATA / 'interactions/instamart_orders.csv')
dineout_bookings = pd.read_csv(DATA / 'interactions/dineout_bookings.csv')
print(f'Food: {len(food_orders)}, Instamart: {len(instamart_orders)}, Dineout: {len(dineout_bookings)}')
```

```python
from bgi_trident.graph.builder import ConsumptionGraphBuilder, InteractionData
data = InteractionData(food_orders=food_orders, instamart_orders=instamart_orders, dineout_bookings=dineout_bookings, users=users, restaurants=restaurants, products=products, venues=venues)
builder = ConsumptionGraphBuilder(data).build()
print('Node counts:', builder.node_counts)
print('Edge counts:', builder.edge_counts)
```

```python
pyg_data = builder.to_pyg()
print('PyG HeteroData:', pyg_data)
```
