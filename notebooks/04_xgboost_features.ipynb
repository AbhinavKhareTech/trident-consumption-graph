```python
import pandas as pd
from pathlib import Path
from bgi_trident.graph.xgboost.model import XGBoostFeatureExtractor
DATA = Path('../src/data')
extractor = XGBoostFeatureExtractor(pd.read_csv(DATA/'interactions/food_orders.csv'), pd.read_csv(DATA/'interactions/instamart_orders.csv'), pd.read_csv(DATA/'interactions/dineout_bookings.csv'))
features = extractor.extract_food_features()
print(f'Rows: {len(features)}, Cols: {len(features.columns)}')
features.head()
```
